```python
#Code
#Creating a new column rating_ratio = rating_numerator/rating_denominator
tweets_archive_clean['rating_ratio'] = tweets_archive_clean['rating_numerator']/tweets_archive_clean['rating_denominator']
```

```python
#Evaluating rating_ratio column in tweets_archive_clean dataframe 
tweets_archive_clean.groupby('rating_ratio').size()
```

```python
#Create a new column called as rating_ratio in the tweets_archive_clean dataframe and setting all rows where ratio is
#greater than or equal to 1.5 to 1.0
for item in range(len(tweets_archive_clean)):
    if (tweets_archive_clean.loc[item, "rating_ratio"] >= 1.5):
        tweets_archive_clean.loc[item, "rating_ratio"] = 1.0
```

```python
#Test
tweets_archive_clean.groupby('rating_ratio').size()
```

```python
#Code
#Concatenate values from all 4 columns doggo, fluffer, pupper and puppo into one column. 
tweets_archive_clean['dog_type_concat'] = tweets_archive_clean.doggo + tweets_archive_clean.floofer + tweets_archive_clean.pupper + tweets_archive_clean.puppo
```

```python
tweets_archive_clean.info()
```

```python
#Use groupby to determine different categories 
tweets_archive_clean.groupby('dog_type_concat').size()
```

```python
#Use string menthod to find a specific value and categorize.
#Code to convert dog_type_concat column to appropriate dog_stage
for tweet in range(len(tweets_archive_clean)):
    if(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'NoneNoneNonepuppo'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'puppo'
    elif(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'NoneNonepupperNone'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'pupper'
    elif(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'NoneflooferNoneNone'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'floofer'
    elif(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'doggoNoneNonepuppo'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'doggo, puppo'
    elif(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'doggoNonepupperNone'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'doggo, pupper'
    elif(tweets_archive_clean.loc[tweet,'dog_type_concat'] == 'doggoflooferNoneNone'):
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'doggo, floofer'
    else:
        tweets_archive_clean.loc[tweet,'dog_type_concat'] = 'None'
```

```python
#Test
tweets_archive_clean.groupby('dog_type_concat').size()
```
